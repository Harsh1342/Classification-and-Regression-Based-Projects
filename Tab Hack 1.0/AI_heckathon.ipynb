```python
import numpy as np
import pandas as pd

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from tqdm.auto import tqdm
from sklearn.metrics import f1_score
```

```python
train_df = pd.read_csv("Week5_train.csv")
```

```python
# print(test_df.shape)
print(train_df.shape)
```

```python
train_df.head()
```

```python
train_df.columns
```

### Dropping all the columns that contains text

```python
train_df.drop(["education"],axis = 1, inplace = True)
```

```python
train_df.drop(["id"],axis = 1, inplace = True)
train_df.nunique()
```

```python
train_df.info()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

train_df["Working_class"] = le.fit_transform(train_df["Working_class"])
train_df["marital_status"] = le.fit_transform(train_df["marital_status"])
train_df["Occupation"] = le.fit_transform(train_df["Occupation"])
train_df["relationship"] = le.fit_transform(train_df["relationship"])
train_df["race"] = le.fit_transform(train_df["race"])
train_df["gender"] = le.fit_transform(train_df["gender"])
train_df["native_country"] = le.fit_transform(train_df["native_country"])
train_df.head()
```

```python
train_df.shape
```

```python
Y = train_df["earning"]
X = train_df.drop("earning",axis = 1,inplace = True)

```

```python
X = train_df
print(X.shape)
```

```python
train_df.head()
```

### Performing the Train Test Split

```python
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size = 0.1)
```

```python
print(y_train.value_counts())
```

```python
y_val.value_counts()
```

### Training

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1200,
 min_samples_split= 5,
 min_samples_leaf= 2,
 max_features= 'sqrt',
 max_depth= 20,
 bootstrap= True)
rf.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
from catboost import CatBoostClassifier
lr = CatBoostClassifier()
lr.fit(X_train , y_train)
```

```python
y_pred_rf = rf.predict(X_val)
y_pred_dt = dt.predict(X_val)
y_pred_gnb = gnb.predict(X_val)
y_pred_lr = lr.predict(X_val)
```

```python
score_rf = f1_score(y_val, y_pred_rf)
score_dt = f1_score(y_val,y_pred_dt)
score_gnb = f1_score(y_val,y_pred_gnb)
score_lr = f1_score(y_val,y_pred_lr)
```

```python
print(score_rf , score_dt , score_gnb , score_lr)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
print(confusion_matrix(y_val , y_pred_lr))
print(accuracy_score(y_val , y_pred_lr))
```
